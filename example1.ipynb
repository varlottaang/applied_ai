# 1. facebook/distilbart-cnn-12-6
Model used https://huggingface.co/sshleifer/distilbart-cnn-12-6

```python
from transformers import pipeline
import os

# Model used https://huggingface.co/sshleifer/distilbart-cnn-12-6
# Open and read the article
f = open("article_5.txt", "r", encoding="utf8")
to_tokenize = f.read()

# Initialize the HuggingFace summarization pipeline
summarizer = pipeline("summarization")
summarized = summarizer(to_tokenize, min_length=256, max_length=600,  truncation=True)

# Print summarized text
print(summarized)
```

# 2. facebook/bart-large-cnn
https://huggingface.co/facebook/bart-large-cnn

```python
from transformers import pipeline
import os

# Open and read the article
f = open("article_5.txt", "r", encoding="utf8")
to_tokenize = f.read()

# Initialize the HuggingFace summarization pipeline
# tokenize to cut the text in chunks, truncation=true to cut if the text seq exceed the
# the max_value. Changing min_length the length of the summary should change :)
# Model used
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
summarized = summarizer(to_tokenize,  min_length=256, max_length=600, truncation=True)

# Print summarized text
print(summarized)
```

# 3. Falconsai/text_summarization
https://huggingface.co/Falconsai/text_summarization

```python
from transformers import pipeline
import os

# Open and read the article
f = open("article_5.txt", "r", encoding="utf8")
to_tokenize = f.read()

# Initialize the HuggingFace summarization pipeline
# tokenize to cut the text in chunks, truncation=true to cut if the text seq exceed the
# the max_value. Changing min_length the length of the summary should change :)

summarizer = pipeline("summarization", model="Falconsai/text_summarization")
summarized = summarizer(to_tokenize,  min_length=100, max_length=300, truncation=True)

# Print summarized text
print(summarized)
```
